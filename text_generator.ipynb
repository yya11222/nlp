```python
from transformers import pipeline

generator = pipeline("text-generation", model= "gpt2")

prompt = "Once upon a time in Kathmandu,"

result = generator(
    prompt,
    max_new_tokens = 50,
    num_return_sequences = 2,

    do_sample = True

)
print(result[0]["generated_text"])
print(result[1]["generated_text"])
```
